```python
#initialize empty lists
country = []
pop2021 = []
area = []

# define the number of iterations of our FOR loop by checking how many elements are in the retrieved result set
# this is equivalent but more robust than just explicitly defining # iterations
#num_iter = len(soup.select("td.titleColumn a"))

num_iter = (1,209)  #number of countries on list/rows in chart

country_list = soup.select("#popTable > div.container.is-fluid.m-0 > div > div > div > div.datatable-container.undefined > table > tbody > tr:nth-child(i) > td:nth-child(2) > a")[0]
population_list = soup.select("#popTable > div.container.is-fluid.m-0 > div > div > div > div.datatable-container.undefined > table > tbody > tr:nth-child(i) > td:nth-child(4)")[0]
area_list = soup.select("#popTable > div.container.is-fluid.m-0 > div > div > div > div.datatable-container.undefined > table > tbody > tr:nth-child(i) > td:nth-child(6)")[0]
# iterate through the result set and retrive all the data


for i in range(num_iter):
    country.append(country_list[i].get_text())
    pop2021.append(population_list[i].get_text())
    area.append(area_list[i].get_text())

```

```python
country = []
pop2021 = []
area = []
density = []

for i in range(1,210):
    country_soup = soup.select('#popTable > div.container.is-fluid.m-0 > div > div > div > div.datatable-container.undefined > table > tbody > tr:nth-child(' + str(i) + ') > td:nth-child(2) > a')[0]
    pop_soup = soup.select('#popTable > div.container.is-fluid.m-0 > div > div > div > div.datatable-container.undefined > table > tbody > tr:nth-child(' + str(i) + ') > td:nth-child(4)')[0]
    area_soup = soup.select('#popTable > div.container.is-fluid.m-0 > div > div > div > div.datatable-container.undefined > table > tbody > tr:nth-child(' + str(i) + ') > td:nth-child(6)')[0]
    density_soup = soup.select('#popTable > div.container.is-fluid.m-0 > div > div > div > div.datatable-container.undefined > table > tbody > tr:nth-child(' + str(i) + ') > td:nth-child(7)')[0]
    
    country.append(country_soup.get_text()) 
    pop2021.append(pop_soup.get_text())
    area.append(area_soup.get_text())
    density.append(density_soup.get_text())
```
