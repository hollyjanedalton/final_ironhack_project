```python
def get_distance_from_lat_lon(lat1, lon1, lat2, lon2):
    import math
    from lat_lon_parser import parse
    R = 6373.0                              # radius of the Earth
    lat1r = math.radians(parse(str(lat1)))  # coordinates Home
    lon1r = math.radians(parse(str(lon1)))
    lat2r = math.radians(parse(str(lat2))) # coordinates Sidney Opera House
    lon2r = math.radians(parse(str(lon2)))
    dlon = lon2r - lon1r                   # difference in coordinates
    dlat = lat2r - lat1r
    a = math.sin(dlat / 2)**2 + math.cos(lat1r) * math.cos(lat2r) * math.sin(dlon / 2)**2 # Haversine formula
    c = 2 * math.atan2(math.sqrt(a), math.sqrt(1 - a))
    distance = R * c
    return distance
get_distance_from_lat_lon(52.168413,4.529004,"33°51\'37.2\"S","151°12\'54.5\"E" )
print(distance)
```
