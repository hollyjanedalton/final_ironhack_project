### Using regex to select the opening hours of each row


-Keep in mind: 18% of rows have been filled with "Unknown" where there were NaNs

-Will go back later to fill these with their literal "nearest neighbor" 

```python

looking for 1 digit, : 2 digits (a space) and then AM 
    create a pattern as a string
pattern= regex string 
if add [0] re.findall
```

```python
import re 
re.findall('AM', sb21_new.iloc[0]['open_hours'])[0]

#the first [1] is what will be iterated through (the index of the rows)
#for i = 0 to 28289, then put all those results in a list and assign that list to a new column
```

```python
re.findall('AM', sb21_new.iloc[0]['open_hours'])[0]
#this is showing, inside the sb21_new database, locate the first row, 
#within that row, the column "open_hours",
#within that row and column, the first part of the dictionary? 
```

```python
re.findall('(0?[1-9]|1[0-2]):([0-5]\d)\s?', sb21_new.iloc[1975]['open_hours'])[0]  

#without the AM/PM, later found out this is necessary because some stores open 12pm/12am (24 hour)
# https://regexland.com/regex-time-am-pm/   used this site, thanks to Erin! 
```

```python
re.findall('(0?[1-9]|1[0-2]):([0-5]\d)\s?(AM|PM)', sb21_new.iloc[1975]['open_hours'])[0]  
#with the AM/PM
```

```python
#next: iterate through each row using a FOR loop and select this information and put inside a list 
#then create a new column with this list called "opening_time"
```

```python
list_hours=[]

for i in range(28289):
    time = re.findall('([0-2]?[1-9]|1[0-2]):([0-5]\d)\s?(AM|PM)', sb21_new.iloc[i]['open_hours'])[0]
    list_hours.append(time)
```

```python
len(list_hours)
```

```python
sb21_new['opening_time'] = list_hours
sb21_new
```

```python
def get_distance_from_lat_lon(lat1, lon1, lat2, lon2):
    import math
    from lat_lon_parser import parse
    R = 6373.0                              # radius of the Earth
    lat1r = math.radians(parse(str(lat1)))  # coordinates Home
    lon1r = math.radians(parse(str(lon1)))
    lat2r = math.radians(parse(str(lat2))) # coordinates Sidney Opera House
    lon2r = math.radians(parse(str(lon2)))
    dlon = lon2r - lon1r                   # difference in coordinates
    dlat = lat2r - lat1r
    a = math.sin(dlat / 2)**2 + math.cos(lat1r) * math.cos(lat2r) * math.sin(dlon / 2)**2 # Haversine formula
    c = 2 * math.atan2(math.sqrt(a), math.sqrt(1 - a))
    distance = R * c
    return distance
get_distance_from_lat_lon(52.168413,4.529004,"33°51\'37.2\"S","151°12\'54.5\"E" )
print(distance)
```
