```python
sb3=pd.read_csv('starbucks_2018_11_12.csv')
sb3

#Starbucks December 2018 dataset
#source: https://data.world/data-hut/starbucks-location-dataset
```

```python
#drop columns for NaNs/unnecessary information

new = sb3[['name', 'url', 'street_address', 'city', 'state', 'zip_code', 'country', 'latitude', 'longitude', 'open_hours']].copy()
new
```

```python
percent_missing = new.isnull().sum() * 100 / len(new)
percent_missing   #decide later what to do with nulls in city, state, zipcode, open_hours
```

```python
#show me all the rows where state = NaN
#I'm guessing these are stores that are located outside the US

state_NaN = new[new['state'].isna()]
state_NaN.query("country == 'US'")  #yes, can see there aren't any rows from the US
```

```python
#replace null values in state column with "Not in US"

new["state"].fillna("Not in US", inplace = True)
```

```python
new['state'].isnull().sum()
```

```python
#But there also 15,000 stores that have a "state", but are not in the US. 

#Don't think state will be a very important feature, will possibly be dropped later.
#Not worth spending a lot of time on the "state" column. Leave it as is for now. 

state_df = new.query("country != 'US'")
state_df
```
