```python
import pandas as pd
import numpy as np
```

```python
sb3=pd.read_csv('starbucks_2018_11_12.csv')
sb3

#source: https://data.world/data-hut/starbucks-location-dataset
```

```python
percent_missing = sb3.isnull().sum() * 100 / len(sb3)
percent_missing   

#there are a lot of NaNs but most of them are in fields that I'm not interested in. 
#It's a bit worrisome though
```

```python
#drop columns for NaNs/unnecessary information

new = sb3[['name', 'url', 'street_address', 'city', 'state', 'zip_code', 'country', 'latitude', 'longitude', 'open_hours']].copy()
new
```

```python
percent_missing = new.isnull().sum() * 100 / len(new)
percent_missing   #decide later what to do with nulls in city, state, zipcode, open_hours
```

```python
#show me all the rows where state = NaN
#I'm guessing these are stores that are located outside the US

state_NaN = new[new['state'].isna()]
state_NaN.query("country == 'US'")  #yes, can see there aren't any rows from the US
```

```python
#replace null values in state column with "Not in US"

new["state"].fillna("Not in US", inplace = True)
```

```python
new['state'].isnull().sum()
```

```python
percent_missing = new.isnull().sum() * 100 / len(new)
percent_missing   #still have null values in city, zip code, open_hours 
```

```python
city_NaN = new[new['city'].isna()]   #looks like the NaNs are in either Egypt, Saudia Arabia, and Myanmar
city_NaN

#only 12 NaNs
#don't want to delete these rows, also don't think knowing which city these stores are in is very important. 
```

```python
#replace NaN values in "city" column with "Unknown"

new["city"].fillna("Unknown", inplace = True)
new
```

```python
percent_missing = new.isnull().sum() * 100 / len(new)
percent_missing
```

```python
zip_NaN = new[new['zip_code'].isna()]   #looks like the NaNs are in either Egypt, Saudia Arabia, and Myanmar
zip_NaN   #7% of stores with unknown zipcode. don't expect zip code to be a very important feature
```

```python
new["zip_code"].fillna("Unknown", inplace = True)
new
```

```python
hours_NaN = new[new['open_hours'].isna()]   #15% of stores don't show opening hours
hours_NaN
```

```python
new["open_hours"].fillna("Unknown", inplace = True)
new
```

```python
percent_missing = new.isnull().sum() * 100 / len(new)
percent_missing   #no more NaNs!! 
```
